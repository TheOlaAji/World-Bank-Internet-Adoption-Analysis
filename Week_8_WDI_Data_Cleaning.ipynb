```python
internet_indicators = df[df["Indicator Name"].str.contains("internet", case=False, na=False)]

internet_indicators["Indicator Name"].unique()
```

```python
selected_indicators = [
    "Individuals using the Internet (% of population)",
    "Individuals using the Internet, female (% of female population)",
    "Individuals using the Internet, male (% of male population)",
    "Secure Internet servers (per 1 million people)"
]

filtered_df = df[df["Indicator Name"].isin(selected_indicators)]

filtered_df.head()
```

```python
selected_countries = [
    "Nigeria",
    "United States",
    "United Kingdom",
    "China",
    "India",
    "Japan",
    "Brazil",
    "South Africa",
    "Germany",
    "Canada"
]

country_df = filtered_df[filtered_df["Country Name"].isin(selected_countries)]

country_df.shape
```

```python
long_df = country_df.melt(
    id_vars=["Country Name", "Country Code", "Indicator Name", "Indicator Code"],
    var_name="Year",
    value_name="Value"
)

long_df.head()
```

```python
long_df.shape
```

```python
long_df["Year"] = long_df["Year"].astype(int)

long_df.dtypes
```

```python
# Remove rows with missing values
clean_df = long_df.dropna(subset=["Value"])

# Check the new shape
clean_df.shape
```

```python
clean_df.to_csv(
    r"C:\Users\HomePC\Desktop\Week 8 AnalystLab\Cleaned_Internet_Indicators.csv",
    index=False
)

print("Dataset saved successfully!")
```

```python
internet_users = clean_df[
    clean_df["Indicator Name"] == "Individuals using the Internet (% of population)"
]

internet_users.head()
```

```python
internet_users.shape
```

```python
average_internet = (
    internet_users.groupby("Country Name")["Value"]
    .mean()
    .sort_values(ascending=False)
)

average_internet
```
